## Aufgabe 9

### a) Test

```python
from project_b4.simulation.detector.exp_angle_dist import ExponentialDeltaAngleDistribution 
```

```python
EDAD = ExponentialDeltaAngleDistribution(k = 1.)
```

```python
x_linspace = np.linspace(-3/2*np.pi,3/2*np.pi,10000)
x_float = np.pi/2
```

```python
plt.plot(x_linspace,EDAD.pdf(x_linspace),'-',label='pdf from numpy linspace')
plt.plot(x_float,EDAD.pdf(x_float),'o',label='pdf from float')
plt.legend()
plt.show()
```

### b) Test

```python
x_linspace = np.linspace(-3/2*np.pi,3/2*np.pi,10000)
x_float = np.pi/2
```

```python
plt.plot(x_linspace,EDAD.cdf(x_linspace),'-',label='cdf from numpy linspace')
plt.plot(x_float,EDAD.cdf(x_float),'o',label='cdf from float')
plt.legend()
plt.show()
```

### c) Test

```python
u_linspace = np.linspace(0,1,10000)
u_float = 0.5
```

```python
plt.plot(u_linspace,EDAD.ppf(u_linspace),'-',label='ppf from numpy linspace')
plt.plot(u_float,EDAD.ppf(u_float),'o',label='ppf from float')
plt.legend()
plt.show()
```
